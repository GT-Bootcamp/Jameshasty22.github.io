```python
import pandas as pd
```

```python
data = "cities.csv"
```

```python
weather_df = pd.read_csv(data)
```

```python
weather_df.head()
```

```python
import plotly.tools as tl
import plotly.express as px
import plotly.offline as py
from plotly.offline import iplot
import plotly.graph_objects as go
from plotly.subplots import make_subplots

#World Map with Max Temp and Humidity
fig = px.scatter_mapbox(weather_df, title="World Map with Max Temperatures",  lat="Lat", lon="Lng", hover_name="Country", hover_data=["Max Temp", "Humidity"],
                        color_discrete_sequence=["fuchsia"], zoom=3, height=800)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(title_x=0.5,
    title_y=0.95,
    mapbox={"style": "carto-positron", "center": {"lon": 2, "lat" : 47}, "zoom": 4.8},
    margin={"l": 0, "r": 0, "b": 0, "t": 0})
fig.show()
fig.write_html('world_map_heat.html', auto_open=True)
```

```python
fig = px.density_mapbox(weather_df, lat='Lat', lon='Lng', z='Max Temp', radius=10,
                        center=dict(lat=0, lon=180), zoom=0,
                        mapbox_style="open-street-map")
fig.show()
fig.write_html('world_map_Max Temp.html', auto_open=True)
```

```python
fig = px.density_mapbox(weather_df, lat='Lat', lon='Lng', z='Cloudiness', radius=10,
                        center=dict(lat=0, lon=180), zoom=0,
                        mapbox_style="stamen-terrain")
fig.show()
fig.write_html('world_map_Cloudiness.html', auto_open=True)
```

```python
fig = px.scatter_geo(weather_df, lat='Lat', lon='Lng', color="Humidity",
                     hover_name="City", size="Humidity",
                     projection="natural earth")
fig.show()
fig.write_html('world_map_Humidity.html', auto_open=True)
```

```python
fig = px.scatter_geo(weather_df, lat='Lat', lon='Lng', color="Wind Speed",
                     hover_name="City", size="Wind Speed",
                     projection="natural earth")
fig.show()
fig.write_html('world_map_Wind Speed.html', auto_open=True)
```
